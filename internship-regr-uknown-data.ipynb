<h3> Normalize data divide there value on there max values

```python
x_max = train_x.max()
train_x_norm = train_x / x_max
for col in train_x_norm.columns:
    print(f"#{col} min: {round(train_x_norm[col].min(),3)}, max: {round(train_x_norm[col].max(),3)}, mean:{round(train_x_norm[col].mean(),3)}")
```

```python
y_max = train_y.max()
train_y_norm = train_y / y_max
train_y_norm.describe()
```

```python
train_x_norm, test_x_norm, train_y_norm, test_y_norm = train_test_split(train_x_norm, train_y_norm, test_size=0.10, random_state=18)
train_x, test_x, train_y, test_y = train_test_split(train_x, train_y, test_size=0.10, random_state=18)
print(train_x_norm.shape, test_x_norm.shape, train_y_norm.shape, test_y_norm.shape)
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
%%time
from sklearn.ensemble import ExtraTreesRegressor
model_5 = ExtraTreesRegressor(n_estimators=15, max_depth=5, random_state=18, criterion="mse")
model_5.fit(train_x_norm, train_y_norm)
model_5.score(train_x_norm, train_y_norm)
```

```python
%%time
# model_5.score(test_x, test_y)
prediction = model_5.predict(train_x_norm)
print(rmse(predictions=prediction, targets=train_y_norm))
```
